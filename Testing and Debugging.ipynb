```python
import numpy as np
from statistics import mean
import matplotlib.pyplot as plt
```

```python
y = []
for _ in range(12):
    y.append(np.random.rand() * 10)
x = np.arange(0,12)
```

```python
def SmoothedValue(y,smoothing_factor,previous_values):
    y_ = []
    for i in range(previous_values):
        y_.append(y[i])
    print(y_)
    for i in range(previous_values,len(y)):
        current_value = smoothing_factor*y[i]
        addedSum = 0
        for j in range(max(i-1,0),max(i-previous_values,-1),-1):
            addedSum += y_[j]
        print(addedSum)
        y_.append(current_value + (1-smoothing_factor)*(addedSum/previous_values))
    return y_
```

```python
y_ = SmoothedValue(y,0.8,5)
```

```python
%matplotlib inline

plt.plot(x,y,color='red')
plt.plot(x,y_,color='blue')
plt.show()
```
